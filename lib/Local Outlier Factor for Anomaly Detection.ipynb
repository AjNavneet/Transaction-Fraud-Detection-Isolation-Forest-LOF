# What is LOF (Local Outlier Factor)?

LOF is an unsupervised machine learning algorithm that identifies outliers based on local neighborhoods, rather than the entire data distribution. It calculates the extent to which a data point deviates from its local cluster.

The main advantage of using LOF is its ability to identify points that are outliers relative to their local cluster of points. This local context helps in distinguishing anomalies that might not be considered outliers when considering the entire dataset. LOF is particularly useful in detecting anomalies in complex and high-dimensional datasets.

---

### Basic Idea of LOF (Local Outlier Factor)

The core concept behind the Local Outlier Factor (LOF) is rooted in the notion of local density estimation. LOF calculates the density of data points within a local neighborhood, where locality is defined by the k nearest neighbors. The distances to these neighbors are used to estimate the local density of a data point.

By comparing the local density of an object to the local densities of its neighbors, one can identify regions of similar density and, importantly, points that exhibit a significantly lower density than their neighbors. These data points, with much lower densities than their local counterparts, are considered outliers by the LOF algorithm.

The essence of LOF lies in the comparison of the local density of a data point with the densities of its neighboring data points. Outliers, or anomalies, are those data points that stand out due to their substantially lower local density.



---


### Estimating Local Density and Reachability Distance in LOF

In the Local Outlier Factor (LOF) algorithm, the local density of a data point is estimated based on the typical distance at which a point can be "reached" from its neighboring data points. To perform this estimation, LOF employs the concept of the "reachability distance."

The reachability distance is defined as the maximum of two distances:
1. The Euclidean distance between two data points.
2. The k-distance of the point, where k-distance(A) is the distance from the object A to its k-th nearest neighbor.

This concept helps LOF measure the relative density of a data point within its local neighborhood. Data points with lower reachability distances tend to be more isolated from their neighbors and are more likely to be considered outliers.

By comparing the reachability distances of data points and their neighbors, LOF identifies regions with differing densities and highlights data points with reachability distances substantially higher or lower than their neighbors. This approach allows LOF to detect local outliers effectively.

---

### Calculating Reachability Distances and Local Reachability Density in LOF

In the Local Outlier Factor (LOF) algorithm, the calculation of reachability distances and the subsequent determination of Local Reachability Density (LRD) play a crucial role in identifying local outliers.

1. **Reachability Distance**:
   - For data points within a local cluster (inside the circle in the diagram), the k-distance is considered. The k-distance (k-distance(A)) represents the distance from the object A to its k-th nearest neighbor.
   - For data points outside the cluster (in the shaded area), the direct distance between the points is considered. The reachability distance measures how "reachable" a point is from another point.

2. **Calculation of Local Reachability Density (LRD)**:
   - To assess the LRD of a point, reachability distances to all of its k-nearest neighbors are computed.
   - The LRD is a measure of the density of k-nearest points around a given point. It is calculated as the inverse of the sum of all the reachability distances of all k-nearest neighboring points.
   - In the equation, the inverse is used to account for the fact that higher values of reachability distances indicate lower density. Closer points have lower distances and higher densities, so taking the inverse aligns the measure with the concept that higher values represent lower density.

The diagram illustrates how these concepts are applied, helping LOF to differentiate between regions of varying data densities and effectively identify local outliers.

### Interpreting LOF (Local Outlier Factor) Values

The Local Outlier Factor (LOF) provides a valuable metric for assessing the similarity of data points within their local neighborhoods. The interpretation of LOF values is as follows:

- **LOF == 1**: Similar Data Point
  - When the LOF is approximately equal to 1, it suggests that the data point is similar to its neighbors in terms of local density. This indicates that the data point is part of a relatively uniform density cluster.

- **LOF < 1**: Inlier (Data Point Inside the Density Cluster)
  - LOF values less than 1 imply that the data point is an inlier, meaning it is similar to its neighbors and located inside the local density cluster. Such data points are not considered outliers.

- **LOF > 1**: Outlier (Anomalous Data Point)
  - When the LOF exceeds 1, it signifies that the data point's local density is substantially different from its neighbors. In this context, the data point is classified as an outlier, as it deviates from the surrounding density pattern.

In summary, LOF values provide a clear way to distinguish between similar data points, inliers located within the density cluster, and outliers that stand out from their local neighborhood.

These interpretations are valuable for identifying and understanding data points in complex datasets.

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
from numpy import quantile, where, random
```

```python
# Reading the data
transaction_data = pd.read_csv(r"input\credit_card_transactional_data.csv")
transaction_data.tail(10)
```

```python
# Checking data information
transaction_data.info()
```

```python
# Checking class distribution of the target variable
transaction_data['Class'].value_counts()
```

```python
# Plot to visualize class distribution of the target variable
sns.countplot(x='Class',data=transaction_data)
```

```python
# Statistical analysis
transaction_data.describe()
```

```python
# Checking for null values
transaction_data.isnull().sum()
```

```python
# Imputing null values with median
transaction_data = transaction_data.fillna(transaction_data.median())
```

```python
# Rechecking for null values
transaction_data.isnull().sum()
```

```python
# Plotting heatmap to check correlation between variables
sns.heatmap(transaction_data.corr())
```

```python
# Finding correlation between variables
transaction_data.corr()
```

```python
transaction_data['Class'].value_counts()
```

```python
# Calculate the number of fraud transactions
Fraud = transaction_data[transaction_data['Class'] == 1]

# Calculate the number of valid (non-fraud) transactions
Valid = transaction_data[transaction_data['Class'] == 0]

# Calculate the contamination value, which is the ratio of fraud transactions to valid transactions
contamination = len(Fraud) / float(len(Valid))
```

```python
print(contamination)

print("Fraud Class : {}".format(len(Fraud)))

print("Normal Class : {}".format(len(Valid)))
```

```python
contamination
```

```python
# Independent variables
X = transaction_data.drop('Class',axis=1)

# Dependent Variable
y = transaction_data['Class']
```

The number of neighbors considered, (parameter n_neighbors) is typically chosen 1) greater than the minimum number of objects a cluster has to contain, so that other objects can be local outliers relative to this cluster, and 2) smaller than the maximum number of close by objects that can potentially be local outliers. In practice, such informations are generally not available, and taking n_neighbors=20 appears to work well in general.

### Fitting the model and predicting with LOF with our contamination number

```python
# Creating an LOF (Local Outlier Factor) model with specific hyperparameters
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.0018)
```

```python
# Using the LOF model to predict and classify data points in X as outliers (y_pred)
y_pred = lof.fit_predict(X)
```

```python
y_pred
```

### Converting Input dataframe into numpy array for the visualisation of outliers

```python
# Convert the feature set X to a NumPy array
X_array = X.to_numpy()

# Find the indices of data points predicted as outliers by the LOF model
y_index = np.where(y_pred == -1)

# Extract the values of the outliers from X_array
outlier_values = X_array[y_index]
```

```python
outlier_values
```

### Plotting outliers and normal values

```python
# Create a scatter plot of all data points
plt.scatter(X_array[:, 0], X_array[:, 1])

# Highlight outliers in red
plt.scatter(outlier_values[:, 0], outlier_values[:, 1], color='r')

# Display the plot
plt.show()
```

### Fitting the model and predicting with LOF with default contamination number

```python
# Separating the feature set (X) and the target variable (y)
X = transaction_data.drop('Class', axis=1) 
y = transaction_data['Class']  
```

```python
# Initialize a Local Outlier Factor (LOF) model with default hyperparameters
lof = LocalOutlierFactor()
```

#### Here we will be deciding the threshold value and we will be selecting lowest 1 percent of score values as the anomalies.

```python
# Using the LOF model to predict and classify data points in X as outliers (anomaly_scores)
lof.fit_predict(X)

# Extracting the anomaly scores for each data point
anomaly_scores = lof.negative_outlier_factor_
```

```python
anomaly_scores
```

```python
# Calculate a threshold for identifying outliers based on the anomaly scores
threshold = np.quantile(anomaly_scores, 0.01)
print(threshold)
```

```python
# Convert the feature set X to a NumPy array
X_array = X.to_numpy()

# Find the indices of data points with anomaly scores less than or equal to the threshold
index = np.where(anomaly_scores <= threshold)

# Extract the values of the outliers from X_array
outlier_values = X_array[index]
```

```python
# Create a scatter plot of all data points
plt.scatter(X_array[:, 0], X_array[:, 1])

# Highlight outliers in red
plt.scatter(outlier_values[:, 0], outlier_values[:, 1], color='r')

# Display the plot
plt.show()
```
